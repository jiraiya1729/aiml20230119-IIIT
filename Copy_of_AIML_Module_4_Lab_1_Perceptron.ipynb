```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

First we need to define a labeled data set. Let's make a toy dataset.

```python
X = np.array([
    [-2, 4],
    [4, 1],
    [1, 6],
    [2, 4],
    [6, 2]
])
```

Next we need to add a bias term -1 into the data set.

```python
X = np.array([
    [-2, 4, -1],
    [4, 1, -1],
    [1, 6, -1],
    [2, 4, -1],
    [6, 2, -1],
])
```

```python
y = np.array([-1, -1, 1, 1, 1])
```

This small toy data set contains two samples labeled with âˆ’1 and three samples labeled with +1. This means we have a binary classification problem, as the data set contains two sample classes. Lets plot the dataset to see, that is is linearly seperable

```python
for d, sample in enumerate(X):
    # Plot the negative samples
    if d < 2:
        plt.scatter(sample[0], sample[1], s=120, marker='_', linewidths=2)
    # Plot the positive samples
    else:
        plt.scatter(sample[0], sample[1], s=120, marker='+', linewidths=2)

# Print a possible hyperplane, that is seperating the two classes.
plt.plot([-2, 6], [6, 0.5])
```

Finally we can code our Perceptron algorithm using our update rule. To keep it simple, we will linearly loop over the sample set. For larger data sets it makes sence, to randomly pick a sample during each iteration in the for-loop.

image.png

The algorithm is shown in the above image.

```python
def perceptron_algo(X, Y):
    w = np.zeros(len(X[0]))
    eta = 1
    epochs = 10

    for epoch in range(epochs):
        for i, x in enumerate(X):
            if (np.dot(X[i], w) * Y[i]) <= 0:
                w = w + eta * X[i] * Y[i]
    return w
```

```python
def perceptron_algo_plot(X, Y):
    '''
    train perceptron and plot the total loss in each epoch.

    :param X: data samples
    :param Y: data labels
    :return: weight vector as a numpy array
    '''
    w = np.zeros(len(X[0]))
    eta = 1
    n = 30
    errors = []

    for t in range(n):
        total_error = 0
        for i, x in enumerate(X):
            if (np.dot(X[i], w) * Y[i]) <= 0:
                total_error += (np.dot(X[i], w) * Y[i])
                w = w + eta * X[i] * Y[i]
        errors.append(total_error * -1)

    plt.plot(errors)
    plt.xlabel('Epoch')
    plt.ylabel('Total Loss')

    return w
```

Next, we will see the the training error at each epoch.

```python
w = perceptron_algo_plot(X, y)
```

We now plot the decision boundary and check which samples have been classified as positive and which samples have been classified as negative.

```python
for d, sample in enumerate(X):
    # Plot the negative samples
    if d < 2:
        plt.scatter(sample[0], sample[1], s=120, marker='_', linewidths=2)
    # Plot the positive samples
    else:
        plt.scatter(sample[0], sample[1], s=120, marker='+', linewidths=2)

# Add our test samples

plt.scatter(2, 2, s=120, marker='_', linewidths=2, color='yellow')
plt.scatter(4, 3, s=120, marker='+', linewidths=2, color='blue')

# Print the hyperplane calculated by perceptron_sgd()
x2 = [w[0], w[1], -w[1], w[0]]
x3 = [w[0], w[1], w[1], -w[0]]

x2x3 = np.array([x2, x3])
X, Y, U, V = zip(*x2x3)
ax = plt.gca()
ax.quiver(X, Y, U, V, scale=1, color='blue')
```

As you can see, our algorithm has correctly classified all the samples. The decision boundary is shown in blue.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the Sonar dataset

df = pd.read_csv('sonar_csv.csv')

# Display the first few rows of the dataset
print(df.head())

# Extract features (X) and labels (Y)
X = df.values[:, :-1]  # Features
Y_raw = df.values[:, -1]  # Labels ('M' for mines, 'R' for rocks)

# Convert labels to binary: 1 for 'M' (mines), -1 for 'R' (rocks)
Y = np.where(Y_raw == 'Mine', 1, -1)

# Plot the dataset (you may choose two features for visualization)
plt.scatter(X[Y == 1, 0], X[Y == 1, 1], color='r', marker='+', label='Mines')
plt.scatter(X[Y == -1, 0], X[Y == -1, 1], color='b', marker='_', label='Rocks')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
def perceptron_algorithm(X, Y):
    w = np.zeros(X.shape[1] + 1)  # Initialize weights with an additional bias term
    eta = 1
    epochs = 10

    for epoch in range(epochs):
        for i, x in enumerate(X):
            x_with_bias = np.insert(x, 0, 1)  # Add bias term
            if (np.dot(x_with_bias, w) * Y[i]) <= 0:
                w = w + eta * x_with_bias * Y[i]

    return w

def predict(X, weights):
    X_with_bias = np.insert(X, 0, 1)
    return np.sign(np.dot(X_with_bias, weights))

weights = perceptron_algorithm(X, Y)

# Plot the decision boundary

x1_range = np.linspace(np.min(X[:, 0]), np.max(X[:, 0]), 100)
x2_range = - (weights[0] + weights[1] * x1_range) / weights[2]
plt.scatter(X[Y == 1, 0], X[Y == 1, 1], color='r', marker='+', label='Mines')
plt.scatter(X[Y == -1, 0], X[Y == -1, 1], color='b', marker='_', label='Rocks')
plt.plot(x1_range, x2_range, color='g', label='Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
def perceptron_algorithm(X, Y, eta=1, epochs=10):
    w = np.zeros(X.shape[1] + 1)  # Initialize weights with an additional bias term

    for epoch in range(epochs):
        for i, x in enumerate(X):
            x_with_bias = np.insert(x, 0, 1)  # Add bias term
            if (np.dot(x_with_bias, w) * Y[i]) <= 0:
                w = w + eta * x_with_bias * Y[i]

    return w

# Experiment with different learning rates
learning_rates = [0.01, 0.1, 0.5, 1.0, 2.0]
epochs = 50  # Increase the number of epochs for better observation

plt.figure(figsize=(12, 8))

for lr in learning_rates:
    weights = perceptron_algorithm(X, Y, eta=lr, epochs=epochs)

    x1_range = np.linspace(np.min(X[:, 0]), np.max(X[:, 0]), 100)
    x2_range = - (weights[0] + weights[1] * x1_range) / weights[2]

    plt.plot(x1_range, x2_range, label=f'Learning Rate = {lr}')

plt.scatter(X[Y == 1, 0], X[Y == 1, 1], color='r', marker='+', label='Mines')
plt.scatter(X[Y == -1, 0], X[Y == -1, 1], color='b', marker='_', label='Rocks')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

4. Take a toy dataset that is not linearly separable and run the perceptron algorithm on it. What happens? Note your observations. An example is given below.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a toy dataset that is not linearly separable
np.random.seed(42)
X1 = np.random.normal(loc=[1, 1], scale=0.5, size=(50, 2))
X2 = np.random.normal(loc=[-1, -1], scale=0.5, size=(50, 2))
X = np.vstack([X1, X2])
Y = np.hstack([np.ones(50), -np.ones(50)])

# Shuffle the dataset
shuffle_idx = np.random.permutation(len(X))
X = X[shuffle_idx]
Y = Y[shuffle_idx]

# Plot the dataset
plt.scatter(X[Y == 1, 0], X[Y == 1, 1], color='r', marker='+', label='Class 1')
plt.scatter(X[Y == -1, 0], X[Y == -1, 1], color='b', marker='_', label='Class -1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
def perceptron_algorithm(X, Y, eta=1, epochs=10):
    w = np.zeros(X.shape[1] + 1)  # Initialize weights with an additional bias term

    for epoch in range(epochs):
        for i, x in enumerate(X):
            x_with_bias = np.insert(x, 0, 1)  # Add bias term
            if (np.dot(x_with_bias, w) * Y[i]) <= 0:
                w = w + eta * x_with_bias * Y[i]

    return w

# Train the perceptron on the non-linearly separable dataset
weights = perceptron_algorithm(X, Y)

# Plot the dataset and the decision boundary
plt.scatter(X[Y == 1, 0], X[Y == 1, 1], color='r', marker='+', label='Class 1')
plt.scatter(X[Y == -1, 0], X[Y == -1, 1], color='b', marker='_', label='Class -1')

# Plot the decision boundary
x1_range = np.linspace(np.min(X[:, 0]), np.max(X[:, 0]), 100)
x2_range = - (weights[0] + weights[1] * x1_range) / weights[2]
plt.plot(x1_range, x2_range, color='g', label='Decision Boundary')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
