Creating a function for the derivative of $f(x)$ called df(x)

```python
def df(x):
    return 2 * x + 1
```

## Gradient Descent

```python
##########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################

# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)  ## [REQUIRED]
    new_x = previous_x - step_multiplier * gradient  ## [REQUIRED]

    step_size = abs(new_x - previous_x)
    # print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))  ## [REQUIRED]

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
```

## Gradient Descent

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

# Example 4 - Data Viz with 3D Charts

## Minimise $$f(x, y) = \frac{1}{3^{-x^2 - y^2} + 1}$$

Minimise $$f(x, y) = \frac{1}{r + 1}$$ where $r$ is $3^{-x^2 - y^2}$

```python
## TASK-8 : Code the same things as they were done for the first two examples
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function and derivative definitions for 3D case
def f(x, y):
    r = 3**(-x**2 - y**2)
    return 1 / (r + 1)

def df_dx(x, y):
    r = 3**(-x**2 - y**2)
    return 2 * x * np.log(3) * r / (r + 1)**2

def df_dy(x, y):
    r = 3**(-x**2 - y**2)
    return 2 * y * np.log(3) * r / (r + 1)**2

# Gradient Descent for 3D case
def gradient_descent_3d(derivative_func_x, derivative_func_y, initial_guess, multiplier=0.01, precision=0.001, max_iter=300):
    x, y = initial_guess
    x_list = [x]
    y_list = [y]
    z_list = [f(x, y)]

    for _ in range(max_iter):
        x = x - multiplier * derivative_func_x(x, y)
        y = y - multiplier * derivative_func_y(x, y)
        z = f(x, y)

        x_list.append(x)
        y_list.append(y)
        z_list.append(z)

        if abs(z_list[-2] - z_list[-1]) < precision:
            break

    return x, y, x_list, y_list, z_list

# Set up the 3D surface plot
x_3d = np.linspace(-2, 2, 100)
y_3d = np.linspace(-2, 2, 100)
x_3d, y_3d = np.meshgrid(x_3d, y_3d)
z_3d = f(x_3d, y_3d)

# Plot the 3D surface
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x_3d, y_3d, z_3d, cmap='viridis', alpha=0.8)

# Choose initial guess and run gradient descent
initial_guess_3d = (0.5, 0.5)
local_min_3d, _, x_list_3d, y_list_3d, z_list_3d = gradient_descent_3d(df_dx, df_dy, initial_guess_3d)

# Plot the gradient descent path on the surface plot
ax.scatter(x_list_3d, y_list_3d, z_list_3d, color='red', s=50, label='Gradient Descent Path')
ax.set_title('Gradient Descent in 3D')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X, Y)')

plt.show()

print('Local minimum occurs at:', local_min_3d)
print('Number of steps:', len(x_list_3d))
```

### $$RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$

## Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

## $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

## $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

## MSE & Gradient Descent

```python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
```

```python
multiplier = 0.01
thetas = np.array([2.9, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5)

for i in range(1000):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)

    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mse(y_5, thetas[0] + thetas[1]*x_5))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1]*x_5))
```
